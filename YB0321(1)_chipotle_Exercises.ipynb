### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 3. Assign it to a variable called chipo.

```python
chipo=pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv',sep='\t')
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
import seaborn as sns
import matplotlib.pyplot as plt

#top 5 items
top_items=chipo['item_name'].value_counts().head(5).reset_index()
top_items.columns=['item_name','order_count']

plt.figure(figsize=(10,5))
sns.barplot(data=top_items,x='item_name',y='order_count')
plt.title('Top 5 Most Ordered Items')
plt.xlabel('Item Name')
plt.ylabel('Order Counts')
plt.xticks(rotation=45)
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '', regex=False).astype(float)

order_summary = chipo.groupby('order_id').agg({'item_price': 'sum', 'quantity': 'sum'}).reset_index()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=order_summary, x='item_price', y='quantity')
plt.xlabel('Order Price ($)')
plt.ylabel('Total Items Ordered')
plt.title('Items Ordered per Order Price')
plt.grid(True)
plt.show()
```

### Step 7. BONUS: Create a question and a graph to answer your own question.

```python
#주문별 총 가격의 boxplot 살펴보기

order_price = chipo.groupby('order_id')['item_price'].sum().reset_index()

#Boxplot 생성
plt.figure(figsize=(8, 6))
sns.boxplot(y=order_price['item_price'])
plt.ylim(0,100)
plt.ylabel('Order Price ($)')
plt.title('Boxplot of Order Price')
plt.show()
```
