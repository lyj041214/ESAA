```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("movies_train.csv")
```

```python
data.head()
```

# 5. 배급사별 평균 관객 수 (바 그래프)

## bar chart를 통해 평균 관객 수 알아보기

```python
# 5. 배급사별 평균 관객 수 (바 그래프)

# 배급사별 평균 관객 수 계산
avg_box_off_num_by_distributor = data.groupby("distributor")["box_off_num"].mean().sort_values(ascending=False)

# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_box_off_num_by_distributor.index, y=avg_box_off_num_by_distributor.values, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel("distributor")
plt.ylabel("avg_box_off_num")
plt.title("avg_box_off_num_by_distributor")
plt.show()
```

```python
# X의 범위가 너무 커서 tail과 head 50개씩 따로 살펴보기

# 배급사별 평균 관객 수 계산
avg_audience_by_distributor = data.groupby("distributor")["box_off_num"].mean().sort_values(ascending=False)

print(avg_audience_by_distributor.tail(50))
print('\n')
print(avg_audience_by_distributor.head(50))
```

```python
#head
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_box_off_num_by_distributor.head(50).index, y=avg_box_off_num_by_distributor.head(50).values, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel("distributor")
plt.ylabel("avg_box_off_num")
plt.title("avg_box_off_num_by_distributor")
plt.show()
```

```python
#tail
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_box_off_num_by_distributor.tail(50).index, y=avg_box_off_num_by_distributor.tail(50).values, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel("distributor")
plt.ylabel("avg_box_off_num")
plt.title("avg_box_off_num_by_distributor")
plt.show()
```

```python
# 배급사별 평균 관객 수 계산
avg_box_off_num_by_distributor = data.groupby("distributor")["box_off_num"].mean().sort_values(ascending=False)

# 배급사별로 boxplot 그리기 (평균 관객수가 높은 순서대로)
plt.figure(figsize=(12, 6))
sns.boxplot(x="distributor", y="box_off_num", data=data, order=avg_box_off_num_by_distributor.index, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel("distributor")
plt.ylabel("box_off_num")
plt.title("Boxplot of box_off_num by distributor")
plt.show()
```

bar chart로 배급사 별 평균 관객 수는 알 수 있지만 이상치의 존재 유무와 전체 자료의 분포를 파악할 수 없음.
## boxplot을 통해 distributor 범주 별 자료 분포 알아보기

```python
# head(50)
top_50_distributors = avg_box_off_num_by_distributor.head(50).index

# 상위 50개 배급사의 boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x="distributor", y="box_off_num", data=data[data["distributor"].isin(top_50_distributors)],
            order=top_50_distributors, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel("distributor (Top 50)")
plt.ylabel("box_off_num")
plt.title("Boxplot of box_off_num for Top 50 distributors")
plt.show()
```

```python
# tail(50)
bottom_50_distributors = avg_box_off_num_by_distributor.tail(50).index

# 하위 50개 배급사의 boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x="distributor", y="box_off_num", data=data[data["distributor"].isin(bottom_50_distributors)],
            order=bottom_50_distributors, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel("distributor (Bottom 50)")
plt.ylabel("box_off_num")
plt.title("Boxplot of box_off_num for Bottom 50 distributors")
plt.show()
```

# 6. 상영 시간 vs. 관객 수 관계 (산점도)

```python
# 6. 상영 시간 vs. 관객 수 관계 (산점도)

# 산점도 그리기
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data["time"], y=data["box_off_num"], alpha=0.5)

# 그래프 설정
plt.xlabel("time")
plt.ylabel("box_off_num")
plt.title("time vs box_off_num")
plt.grid(True)
plt.show()
```

시각화 결과 상영 시간이 120분-140분 범주에서 관객 수가 가장 많음을 알 수 있다.

```python
# 로그 스케일을 적용: box_off_num의 분포가 너무 넓어서 쏠려보이기 때문

# 관객 수가 0 이상인 값만 필터링 (로그 변환을 위해)
data_p = data[data["box_off_num"] > 0]

# 그래프 그리기
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_p["time"], y=np.log1p(data_p["box_off_num"]), alpha=0.5)

# 그래프 설정
plt.xlabel("time")
plt.ylabel("box_off_num_log")
plt.title("time vs box_off_num")
plt.grid(True)
plt.show()
```

로그 변환을 했지만 유의미한 결과 도출을 위해서는 로그 변환 이전이 좋을 것 같음.

# 장르와 상영 시간, 상영 등급의 관계 알아보기

```python
# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.boxplot(x="genre", y="time", hue="screening_rat", data=data, palette="Set2")

# 그래프 설정
plt.xticks(rotation=45)
plt.xlabel("genre")
plt.ylabel("time")
plt.title("screening_rat")
plt.legend(title="screening_rat", bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="genre", y="time", hue="screening_rat", data=data, estimator=np.mean, palette="coolwarm")

plt.xticks(rotation=45)
plt.xlabel("Genre")
plt.ylabel("Time (minutes)")
plt.title("Average Running Time by Genre (by Screening Rating)")
plt.legend(title="Screening Rating", bbox_to_anchor=(1, 1))
plt.show()
```

장르별 상영 등급과 상영 시간의 관계는 box plot보다 bar plot으로 나타냈을 때 불필요한 정보가 제거됨. 장르별로 아예 존재하지 않는 상영 등급에 주목하여 장르의 특성 파악하기.
